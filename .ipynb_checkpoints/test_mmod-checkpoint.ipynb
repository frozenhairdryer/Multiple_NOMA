```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng
rng = default_rng()
```

```python
A=rng.integers(0,2,1000)
A2=rng.integers(0,4,1000)
#M1=0.3+0.5*(A+1j*A2)*np.exp(-1j*np.pi/4)
#M1=1+0.5*np.exp(1j*2*np.pi*A/4)
#M1=1+A
M1=np.exp(1j*2*np.pi*A/3)


plt.scatter(np.real(M1),np.imag(M1))
plt.xlim(-2,3)
plt.ylim(-2,2)
plt.grid()
```

```python
B=rng.integers(0,4,1000)
M2=np.exp(1j*2*np.pi*(B/4+1/8))



plt.scatter(np.real(M2),np.imag(M2))
```

```python
#M1=M1+0.05*rng.standard_normal(1000)+1j*0.05*rng.standard_normal(1000)
M=M1*M2

plt.scatter(np.real(M),np.imag(M))
plt.scatter(np.real(M2),np.imag(M2))
plt.grid()
```

```python
#noise
#M=M+0.05*rng.standard_normal(1000)+1j*0.05*rng.standard_normal(1000)
Md=M
plt.scatter(np.real(Md),np.imag(Md))
plt.scatter(np.real(M1),np.imag(M1))
plt.grid()
```

```python
C=rng.integers(0,7,1000)
M3=np.exp(1j*2*np.pi*(C/7))
plt.scatter(np.real(M3),np.imag(M3))
```

```python
M=M*M3
plt.scatter(np.real(M),np.imag(M))
```

```python
#Spiral mod
C=rng.integers(1,5,1000)
phi = 2*np.pi*C/4
Ms = (0.2*C+0.5)*np.exp(1j*phi)
plt.scatter(np.real(Ms),np.imag(Ms))
```

```python
Mj=Ms*M2
plt.scatter(np.real(Mj),np.imag(Mj))
```

```python
Mp=Ms*rng.integers(1,5,1000)
plt.scatter(np.real(Mp),np.imag(Mp))
```

```python
ar=0.5
ai=np.pi/4
Cr=2*rng.integers(0,2,1000)-1
Ci=2*rng.integers(0,2,1000)-1

Mlog=np.exp(ar*Cr+1j*ai*Ci)
plt.scatter(np.real(Mlog),np.imag(Mlog))
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.grid()
```

```python
plt.scatter(np.log(np.abs(Mlog)),np.angle(Mlog))
```

```python
ar2=2
ai2=np.pi/3
Cr2=2*rng.integers(0,2,1000)-1
Ci2=2*rng.integers(0,2,1000)-1

Mg=np.exp(ar2*Cr2+1j*ai2*Ci2)

Mall=Mg*Mlog
plt.scatter(np.real(Mall),np.imag(Mall))
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.grid()
```

```python
plt.scatter(np.log(np.abs(Mall)),np.angle(Mall))
```

```python
## hexagonal
t=np.exp(1j*rng.integers(1,5,1000)*np.pi/2)
plt.scatter(np.real(t),np.imag(t))

tn=0.5+0.25*np.exp(1j*rng.integers(1,5,1000)*np.pi/2)
t1=0.5+0.25*np.exp(1j*1*np.pi/2)
```

```python
plt.scatter(np.real(t*tn),np.imag(t*tn))
plt.scatter(np.real(t*t1),np.imag(t*t1),color='orange')
```
